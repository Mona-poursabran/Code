```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
x1 = np.random.rand(5)
y1 = 2.358 * x1 -3
y1 = y1 + np.random.normal(scale = 0.1 , size = y1.shape)

x1 = x1.reshape(-1, 1)
y1 = y1.reshape(-1, 1)
#plt.plot(x1, y1 , 'bo')
```

```python
model_1 = LinearRegression(normalize=True).fit(x1 , y1)
x1_line= np.arange(0, 2, 0.1).reshape(-1, 1)
y1_line = model_1.predict(x1_line)
plt.plot(x1, y1 , 'bo')
plt.plot(x1_line, y1_line)
score1_1 = model_1.score(x1 , y1)
print(score1_1)
```

```python
poly1 = PolynomialFeatures(degree=4 ,include_bias=False)
x1_4 = poly1.fit_transform(x1)

poly1 = PolynomialFeatures(degree=16 , include_bias=False)
x1_16 = poly1.fit_transform(x1)
```

```python
model_1_4 = LinearRegression().fit(x1_4, y1)

x1_line = np.arange(0, 2, 0.1).reshape(-1 , 1)
x1_line_4 = PolynomialFeatures(degree=4 , include_bias= False).fit_transform(x1_line)
y1_line_4 = model_1_4.predict(x1_line_4)

plt.plot(x1, y1, 'bo')
plt.plot(x1_line, y1_line_4 )

score1_4 = model_1_4.score(x1_4, y1)
print(score1_4)
```

```python
model_1_16 = LinearRegression().fit(x1_16, y1)

x1_line = np.arange(0, 1, 0.01).reshape(-1 , 1)
x1_line_16 = PolynomialFeatures(degree=16 , include_bias=False).fit_transform(x1_line)
y1_line_16 = model_1_16.predict(x1_line_16)

plt.plot(x1, y1, 'bo')
plt.plot(x1_line, y1_line_16 )

score1_16 = model_1_16.score(x1_16, y1)
print(score1_16)
```
