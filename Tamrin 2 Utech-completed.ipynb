```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
```

# Datas:

```python
X  = np.arange(0, 200)
Y = 3.24* X**4 + 2.478 * X**3 + 0.5 * X**2 + 1.42 * X - 3.121
Y = Y - np.random.normal(scale =10 , size = Y.shape)
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)
plt.plot(X, Y, 'o',label = 'Datas' )
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

```python
X_poly_2 = PolynomialFeatures(2).fit_transform(X)
X_poly_4 = PolynomialFeatures(4).fit_transform(X)
X_poly_8 = PolynomialFeatures(8).fit_transform(X)
X_poly_16 = PolynomialFeatures(16).fit_transform(X)
```

# Train_Test_Split

```python
X_train , X_test , Y_train, Y_test = train_test_split(X , Y , test_size = 0.5, random_state = 40)
```

# PolynomialFeatures (degree = 2, 4, 8, 16)

```python
poly_2 = PolynomialFeatures(2, include_bias= False)
X_train_p2 = poly_2.fit_transform(X_train)

poly_2_test = PolynomialFeatures(2, include_bias= False)
X_test_p2 = poly_2_test.fit_transform(X_test)
```

```python
poly_4 = PolynomialFeatures(4, include_bias= False)
X_train_p4 =poly_4.fit_transform(X_train)
X_test_p4 = poly_4.fit_transform(X_test)
```

```python
poly_8 = PolynomialFeatures(8, include_bias= False)
X_train_p8 = poly_8.fit_transform(X_train)
X_test_p8 = poly_8.fit_transform(X_test)
```

```python
poly_16 = PolynomialFeatures(16, include_bias= False)
X_train_p16 = poly_16.fit_transform(X_train)
X_test_p16 = poly_16.fit_transform(X_test)
```

# Model = LinearRegresstion

```python
model_1 = LinearRegression(normalize=True).fit(X_train, Y_train)
model_1 =LinearRegression(normalize=True).fit(X_test, Y_test)
print('Train Score = ', model_1.score(X_train, Y_train))
print('Test Score = ', model_1.score(X_test, Y_test))
```

```python
model_2 = LinearRegression(normalize=True).fit(X_train_p2, Y_train)
model_2 = LinearRegression(normalize=True).fit(X_test_p2, Y_test)
print('Train Score pow 2 = ', model_2.score(X_train_p2, Y_train))
print('Test Score pow 2 = ', model_2.score(X_test_p2, Y_test))
```

```python
model_4 = LinearRegression(normalize=True).fit(X_train_p4, Y_train)
model_4 = LinearRegression(normalize=True).fit(X_test_p4, Y_test)

print('Train Score pow 4 = ', model_4.score(X_train_p4, Y_train))
print('Test Score pow 4 = ', model_4.score(X_test_p4, Y_test))
```

```python
model_8 = LinearRegression(normalize=True).fit(X_train_p8, Y_train)
model_8 = LinearRegression(normalize=True).fit(X_test_p8, Y_test)
print('Train Score pow 8 = ', model_8.score(X_train_p8, Y_train))
print('Test Score pow 8 = ', model_8.score(X_test_p8, Y_test))
```

```python
model_16 = LinearRegression(normalize=True).fit(X_train_p16, Y_train)
model_16 = LinearRegression(normalize=True).fit(X_test_p16, Y_test)
print('Train Score pow 16 = ', model_16.score(X_train_p16, Y_train))
print('Test Score pow 16 = ', model_16.score(X_test_p16, Y_test))
```

# MSE

```python
Y_pre = model_2.predict(X_train_p2)
MSE_train_2= np.mean((Y_train - Y_pre)**2) /2

Y_pre_test = model_2.predict(X_test_p2)
MSE_test_2= np.mean((Y_test - Y_pre_test)**2) /2
print (MSE_train_2)
print (MSE_test_2)
```

```python
Y_pre = model_4.predict(X_train_p4)
MSE_train_4= np.mean((Y_train - Y_pre)**2) /2

Y_pre_test = model_4.predict(X_test_p4)
MSE_test_4= np.mean((Y_test - Y_pre_test)**2) /2

print (MSE_train_4)
print (MSE_test_4)
```

```python
Y_pre = model_8.predict(X_train_p8)
MSE_train_8= np.mean((Y_train - Y_pre)**2) /2

Y_pre_test = model_8.predict(X_test_p8)
MSE_test_8= np.mean((Y_test - Y_pre_test)**2) /2

print (MSE_train_8)
print (MSE_test_8)
```

```python
Y_pre = model_16.predict(X_train_p16)
MSE_train_16= np.mean((Y_train - Y_pre)**2) /2

Y_pre_test = model_16.predict(X_test_p16)
MSE_test_16= np.mean((Y_test - Y_pre_test)**2 )/2

print (MSE_train_16)
print (MSE_test_16)
```

```python
line_x= np.array([1, 2, 4, 8, 16]).reshape(-1, 1)
M_train = np.array([MSE_train, MSE_train_2,MSE_train_4,MSE_train_8, MSE_train_16 ]).reshape(-1, 1)
M_test  = np.array([MSE_test, MSE_test_2,MSE_test_4,MSE_test_8, MSE_test_16 ]).reshape(-1, 1)
```

```python
plt.plot(line_x , M_train, 'b-', label='Train')
plt.plot(line_x, M_test, 'm-', label= 'Test')
plt.legend()
plt.grid()
```

# Learning Curve

```python
lr= LinearRegression(normalize=True)
train_sizes, train_score , test_score = learning_curve(lr , X_train , Y_train)
plt.plot(train_sizes , train_score.mean(axis = 1), 'bo-', label = 'Train score pow 1')
plt.plot(train_sizes , test_score.mean(axis = 1), 'co-', label = 'Test score pow 1')
plt.legend()
print ('Train_score = ' , train_score.mean(axis = 1))
print ('Test_score = ' ,  test_score.mean(axis = 1))
print('Train_sizes = ', train_sizes)
```
