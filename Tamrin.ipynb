**Third question:**

```python
import pandas as pd 
from sklearn.neighbors import KNeighborsClassifier
```

```python
data_train = pd.read_csv('MNIST\mnist_train.csv')
```

```python
data_test = pd.read_csv('MNIST\mnist_test.csv')
```

```python
X_train= data_train.drop(columns='label', axis=1)
Y_train = data_train.label
X_test = data_test.drop(columns='label', axis=1)
Y_test = data_test.label
```

```python
model_clf = KNeighborsClassifier(n_neighbors = 3).fit(X_train , Y_train)
```

```python
print('[INFO] Test Score {}'.format(model_clf.score(X_test, Y_test)))
```

```python
predict= model_clf.predict(X_test)
predict
```

```python
print(' Y_predict:', predict)
print('Y_True: ', Y_test)
```
